```python
#Claire Roush
#claireroush@gmail.com
#1/2/17
#The original CSV was data scraped from http://db.desmoinesregister.com with the ParseHub Project
#2013 City of Des Moines Parks and Recreation employee data
#The original CSV had serious formatting errors. This notebook's project is to clean and organize the data, and write to CSV.
#The only relevant (saved) attributes include Name, Title, Yearly Wage, Hourly Rate, Years Employed

import pandas
df = pandas.read_csv("run_results.csv")
cols = ["Name", "Title", "Wage", "Rate", "Years"]
new_df = pandas.DataFrame(columns = cols)

print(df.shape[0])
i=1 #skip the first useless row
j=0 #index into the new df
while i < df.shape[0]-1:
    new_df.loc[j] = [df.loc[i][0], df.loc[i+1][1], df.loc[i+3][1], df.loc[i+4][1], df.loc[i+5][1]]
    print(new_df.loc[j])
    i=i+7
    j=j+1

new_df.to_csv("cleaned_data.csv")

    
```
